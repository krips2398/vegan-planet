```python
df_province = df.groupby(by='province').agg({'name':'nunique'})
```

```python
df_province.reset_index(inplace=True)
```

```python
def func(x, us_):
    return us_[x].lower()
```

```python
df_province.rename(columns={'province':'region','name':'value'},inplace=True)
```

```python
df_province
```

```python
df_province = df_province[df_province['region']!="AU-WA"]
```

```python
df_province = df_province[df_province['region']!="DC"]
```

```python
df_province['region'] = df_province['region'].apply(lambda x: func(x,us_state_abbrev))
```
